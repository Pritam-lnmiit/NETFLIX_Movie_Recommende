## Feature selection part....

```python
movies=data[['id', 'title', 'genre','overview']]
```

```python
movies.head()
```

```python
movies.info()
```

```python
movies['tags']=movies["genre"]+movies["overview"]
```

```python
movies.head(10)
```

```python
new_data=movies.drop(columns=['genre',"overview"])
```

```python
new_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=10000,stop_words="english")
```

```python
cv
```

```python
vector=cv.fit_transform(new_data["tags"].values.astype('U')).toarray()
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
similarity
```

```python
new_data[new_data["title"]=="The Godfather"].index[0]
```

```python
similarity[2]
```

```python
distance=sorted(list(enumerate(similarity[2])),reverse=True,key=lambda vector:vector[1])
for i in distance[0:5]:
    print(new_data.iloc[i[0]].title)
```

```python
def recommend(movies):
    # Get the index of the movie in the DataFrame
    index = new_data[new_data["title"] == movies].index[0]
    
    # Calculate distances and sort them
    distance = sorted(list(enumerate(similarity[index])), reverse=True, key=lambda vector: vector[1])
    
    # Print the titles of the top 5 recommended movies
    for i in distance[1:6]:  # start from 1 to skip the movie itself
        print(new_data.iloc[i[0]].title)
```

```python
recommend("The Godfather")
```
